### .

### Definindo a função

```python
def model_recipe(x_f, x_c):
    norm_lim = [-1, 1]
    flour_lim = [0.5, 1.5]
    chocolate_lim = [0.1, 0.5]

    # Converter
    x_f_converted = np.interp(x_f, flour_lim, norm_lim)
    x_c_converted = np.interp(x_c, chocolate_lim, norm_lim)

    portions = fited_params['Intercept'] + fited_params['Farinha'] * x_f_converted + fited_params['Chocolate'] * x_c_converted
    return round(portions)
```

```python
model_recipe(0.6, 0.1)
```

```python
x_flour = np.linspace(start=0.5, stop=1.5, num=10)
x_flour
```

```python
x_chocolate = np.linspace(start=0.1, stop=0.5, num=10)
x_chocolate
```

### . 

```python
dots = []

for i in x_flour:
    temp = []
    for j in x_chocolate:
        temp.append(model_recipe(i, j))
    
    dots.append(temp)
```
