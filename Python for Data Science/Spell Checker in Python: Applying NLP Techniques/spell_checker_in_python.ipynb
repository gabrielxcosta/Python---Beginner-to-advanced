# Retirando os caracteres não alfábeticos

```python
def split_words(tokens):
    words_list = []
    for token in tokens:
        if token.isalpha():
            words_list.append(token)
    return words_list
```

```python
splited_words = split_words(splited_words)
splited_words
```

```python
def normalization(words):
    norm = []
    for word in words:
        norm.append(word.lower())
    return norm

norm_words = normalization(splited_words)
print('Quantidade de tipos:', norm_words[:5])
```

# Funções para a operação de correção ortográfica

```python
def insert_letters(slices):
    new_words = []
    letters = 'abcdefghijklmnopqrstuvwxyzàáâãèéêìíîòóôõùúûç'
    for L, R in slices:
        for letter in letters:
            new_words.append(L + letter + R)
    return new_words
```

```python
def word_generator(word):
    slices = [(word[:i], word[i:]) for i in range(len(word) + 1)]
    generated_words = insert_letters(slices)
    return generated_words
```

```python
def spell_checker(word):
    generated_words = word_generator(word)
    correct_word = max(generated_words, key=probability)
    return correct_word
```

# Avaliando nosso modelo

```python
def raise_test_data(filename):
    words_test = []
    with open(filename, 'r') as file:
        for row in file:
            correct, incorrect = row.split()
            words_test.append((correct, incorrect))
    return words_test
```

```python
test_data = raise_test_data('palavras.txt')
test_data
```

```python
def evaluator(tests):
    hit = 0
    for correct, incorrect in tests:
        corrected_word = spell_checker(incorrect)
        if corrected_word == correct:
            hit += 1
    score_rate = hit / len(tests)
    print(f'Taxa de acerto de {len(tests)} palavras: {(score_rate * 100):.2f}%')
```

```python
evaluator(test_data)
```

# Incrementando o corretor

```python
def del_char(slices):
    new_words = []
    for L, R in slices:
        new_words.append(L + R[1:])
    return new_words
```

# Adaptando nosso gerador de palavras a função ``del_char()``

```python
def word_generator(word):
    slices = [(word[:i], word[i:]) for i in range(len(word) + 1)]
    generated_words = insert_letters(slices)
    generated_words += del_char(slices)
    return generated_words
```

# Corrigindo os principais erros de digitação

```python
def exchange_letters(slices):
    new_words = []
    letters = 'abcdefghijklmnopqrstuvwxyzàáâãèéêìíîòóôõùúûç'
    for L, R in slices:
        for letter in letters:
            new_words.append(L + letter + R[1:])
    return new_words
```

# Adaptando nosso gerador de palavras a função ``exchange_letters()``

```python
def word_generator(word):
    slices = [(word[:i], word[i:]) for i in range(len(word) + 1)]
    generated_words = insert_letters(slices)
    generated_words += del_char(slices)
    generated_words += exchange_letters(slices)
    return generated_words
```

```python
test_word = 'lígica'
generated_words = word_generator(test_word)
generated_words
```

```python
def reversing_letters(slices):
    new_words = []
    for L, R in slices:
        if len(R) > 1:
            new_words.append(L + R[1] + R[0] + R[2:])
    return new_words
```

# Adaptando nosso gerador de palavras a função ``reversing_letters()``

```python
def word_generator(word):
    slices = [(word[:i], word[i:]) for i in range(len(word) + 1)]
    generated_words = insert_letters(slices)
    generated_words += del_char(slices)
    generated_words += exchange_letters(slices)
    generated_words += reversing_letters(slices)
    return generated_words
```

# Palavras desconhecidas ao vocabulário do corretor

## Adaptando a função ``evaluator()``

```python
def evaluator(tests, vocabulary):
    hit = 0
    unknown = 0
    for correct, incorrect in tests:
        corrected_word = spell_checker(incorrect)
        if corrected_word == correct:
            hit += 1
        else:
            unknown += (correct not in vocabulary)
    score_rate = hit / len(tests)
    unknown_rate = unknown / len(tests)
    print(f'Taxa de acerto de {len(tests)} palavras: {(score_rate * 100):.2f}%')
    print(f'Taxa de palavras desconhecidas é de: {(unknown_rate * 100):.2f}%')
```

```python
evaluator(test_data, norm_words)
```

# Turbinando nosso gerador

```python
def turbocharged_generator(generated_words):
    new_words = []
    for word in generated_words:
        new_words += word_generator(word)
    return new_words
```

```python
generated_words = turbocharged_generator(word_generator('lóiigica'))
'lógica' in generated_words
```

```python
len(generated_words)
```

# Criando um novo corretor ``spell_checker()``

```python
def new_spell_checker(word, vocabulary):
    generated_words = word_generator(word)
    generated_words_turbocharged = turbocharged_generator(generated_words)
    all_words = set(generated_words + generated_words_turbocharged)
    candidates = [word]
    for word in all_words:
        if word in vocabulary:
            candidates.append(word)
    correct_word = max(candidates, key=probability)
    return correct_word
```

```python
new_spell_checker('lóiigica', norm_words)
```

# Testando o nosso novo corretor

## Adaptando a função ``evaluator()``

```python
def evaluator(tests, vocabulary):
    hit = 0
    unknown = 0
    for correct, incorrect in tests:
        corrected_word = new_spell_checker(incorrect, vocabulary)
        unknown += (correct not in vocabulary)
        if corrected_word == correct:
            hit += 1
    score_rate = hit / len(tests)
    unknown_rate = unknown / len(tests)
    print(f'Taxa de acerto de {len(tests)} palavras: {(score_rate * 100):.2f}%')
    print(f'Taxa de palavras desconhecidas é de: {(unknown_rate * 100):.2f}%')
```
