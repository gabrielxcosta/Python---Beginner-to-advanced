```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
plt.rc('figure', figsize=(15, 8))
```

```python
data = pd.read_csv('dataExtras/aluguel.csv', sep=';')
data.head()
```

```python
area = plt.figure()
```

```python
g1 = area.add_subplot(2, 2, 1)
g2 = area.add_subplot(2, 2, 2)
g3 = area.add_subplot(2, 2, 3)
g4 = area.add_subplot(2, 2, 4)
```

```python
g1.scatter(data.Valor, data.Area)
g1.set_title('Valor vs Área')

g2.hist(data.Valor)
g2.set_title('Histograma dos Valores')

data_g3 = data.Valor.sample(100)
data_g3.index = range(data_g3.shape[0])
g3.plot(data_g3)
g3.set_title('Amostra (Valor)')

grupo = data.groupby('Tipo')['Valor']
label = grupo.mean().index
values = grupo.mean().values
g4.bar(label, values)
g4.set_title('Valor médio por tipo')
```
